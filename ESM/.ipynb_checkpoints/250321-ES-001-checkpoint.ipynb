```python


from sklearn.preprocessing import LabelEncoder

# LabelEncoder 객체 생성
emotion_encoder = LabelEncoder()
situation_encoder = LabelEncoder()

# ✅ 학습 데이터 (train_df)에서 Label Encoding 학습
train_df["emotion_encoded"] = emotion_encoder.fit_transform(train_df["em_code"])
train_df["situation_encoded"] = situation_encoder.fit_transform(train_df["si_code"])


# 확인
print("Train size:", train_df.shape)
```

```python
train_df.head()
```

```python
import re

def preprocess_text(text):
    # 감정 분석에 중요한 특수문자 유지
    important_chars = "!?,~ㅠㅋ^…"
    text = ''.join([char if char in important_chars or char.isalnum() else ' ' for char in text])

    # 연속된 특수문자 축약 (!!!!! → !, ㅋㅋㅋ → ㅋㅋ)
    text = re.sub(r'([!?.])\1+', r'\1', text)
    text = re.sub(r'(ㅋ)\1+', r'\1\1', text)  # ㅋㅋㅋㅋㅋ → ㅋㅋ
    text = re.sub(r'(ㅠ)\1+', r'\1\1', text)  # ㅠㅠㅠㅠ → ㅠㅠ

    return text.strip()


train_df["text"] = train_df["text"].apply(preprocess_text)
```

```python
import random
from tqdm import tqdm
from konlpy.tag import Okt
import pandas as pd
from collections import Counter

# 형태소 분석기
okt = Okt()

#  형태소 분석 후 단어 치환
def simple_eda(text, alpha_sr=0.1):
    words = okt.morphs(text)
    n_sr = max(1, int(len(words) * alpha_sr))
    for _ in range(n_sr):
        idx = random.randint(0, len(words) - 1)
        words[idx] = words[idx]  # 실제로는 동의어 사전 등과 결합 가능
    return ''.join(words)

# 증강이 필요한 감정 클래스 파악
emotion_counts = Counter(train_df["emotion_encoded"])
max_count = max(emotion_counts.values())
target_classes = [cls for cls, count in emotion_counts.items() if count < max_count]

# 증강 실행
augmented_rows = []
for emotion in target_classes:
    subset = train_df[train_df["emotion_encoded"] == emotion]
    needed = max_count - len(subset)
    samples = subset.sample(n=needed, replace=True, random_state=42)
    for _, row in tqdm(samples.iterrows(), total=needed, desc=f"Augmenting emotion {emotion}"):
        augmented_text = simple_eda(row["text"])
        augmented_rows.append({
            "text": augmented_text,
            "emotion_encoded": row["emotion_encoded"],
            "situation_encoded": row["situation_encoded"]
        })

# 증강 데이터프레임 생성 및 병합
df_augmented = pd.DataFrame(augmented_rows)
train_df_augmented = pd.concat([train_df, df_augmented], ignore_index=True)

# 확인
print("📊 증강 후 감정 분포:", Counter(train_df_augmented["emotion_encoded"]))
```

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

# ✅ 오버샘플링 대상: 텍스트 + 감정/상황 라벨 쌍
X = train_df[["text"]]
y = train_df[["emotion_encoded", "situation_encoded"]]

# 문자열 컬럼(X)은 numpy 배열로 변환해야 함
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = ros.fit_resample(X.values, y)

# ✅ 오버샘플링 결과 DataFrame 생성
train_df_resampled = pd.DataFrame({
    "text": X_resampled.flatten(),
    "emotion_encoded": y_resampled["emotion_encoded"],
    "situation_encoded": y_resampled["situation_encoded"]
})

# ✅ 감정+상황 조합의 분포 확인
combo = list(zip(train_df_resampled["emotion_encoded"], train_df_resampled["situation_encoded"]))
print("\n📊 오버샘플링 후 (감정, 상황) 조합 분포:\n", Counter(combo))
```

```python
from sklearn.model_selection import train_test_split

# 80:20 비율로 train/validation split
train_df, val_df = train_test_split(
    train_df, 
    test_size=0.2,
    random_state=42,
    stratify=train_df[['emotion_encoded', 'situation_encoded']]
)

# 분할된 데이터 크기 확인
print(f"Train data shape: {train_df.shape}")
print(f"Validation data shape: {val_df.shape}")
```

```python
import random
import pandas as pd

# 텍스트 템플릿 조각
long_starters = [
    "요즘 너무 지치고 아무것도 하고 싶지 않아요. 아침에 눈을 떠도 하루를 시작하는 게 고통스럽고, 뭔가를 해보려는 의지도 점점 사라져요.",
    "불면증 때문에 밤새 뒤척이다가 결국 아침을 맞이해요. 새벽까지 한숨도 못 자고 머릿속은 생각으로 가득 차 있어요.",
    "감정이 자꾸 무뎌지고 있어요. 누군가와 있어도 외롭고, 예전처럼 즐겁던 일들이 이제는 아무런 의미가 없어요.",
    "사람들 사이에 있으면 숨이 막히고, 모든 말이 비수처럼 느껴져요. 그냥 혼자 있는 게 더 편할 때가 많아졌어요.",
    "이유 없이 눈물이 나고, 모든 게 무의미하게 느껴져요. 웃고 있어도 속은 시끄럽고, 말 한 마디조차 하기 힘들어요.",
]

long_middles = [
    " 머릿속이 너무 복잡해서 아무 생각도 정리가 안 되고, 무엇부터 시작해야 할지 모르겠어요. 누구에게 털어놓기에는 나조차 내 마음을 잘 모르겠어요.",
    " 하루 종일 침대에 누워만 있고 싶은데, 그렇게 보내고 나면 또 스스로가 한심하게 느껴져요. 이런 날들이 반복되면서 자존감도 바닥을 치고 있어요.",
    " 주변 사람들은 다들 잘 살아가는 것처럼 보여서 더 외롭고, 나만 이렇게 멈춰 있는 느낌이 들어요. 그런 생각이 들수록 더 나락으로 빠져들어요.",
    " 스스로를 다그쳐도 달라지는 건 없고, 오히려 더 괴로워져요. 그 누구도 이해해주지 못할 것 같다는 생각이 들어서 입을 닫게 돼요.",
]

long_endings = [
    " 그래서 요즘은 그냥 하루를 버틴다는 느낌으로 살고 있어요. 가끔은 내가 사라졌으면 좋겠다는 생각도 들지만, 그마저도 겁이 나요.",
    " 어떻게든 이 상황에서 벗어나고 싶은데 방법을 모르겠어요. 그냥 괜찮다고 말하는 것도 이제는 지쳤어요.",
    " 나도 예전처럼 웃고 싶은데, 마음이 따라주질 않아요. 이렇게 살아가는 게 맞는 건지 혼란스럽기만 해요.",
    " 무언가를 바꾸고 싶어도, 그 시작조차 막막하게 느껴져요. 스스로를 돌보는 법을 잊어버린 기분이에요.",
]

# 감정 라벨 (랜덤 추론용)
emotion_pool = [
    ("EM02", "슬픔 (Sadness)"),
    ("EM05", "불안 (Anxiety)"),
    ("EM09", "절망감 (Despair)"),
    ("EM11", "우울 (Depression)/외로움"),
    ("EM08", "수치심 (Shame)"),
    ("EM06", "공포 (Fear)")
]

# 상황 카테고리 순환 리스트
situation_categories = [
    ("SI04", "정신 건강 문제"),
    ("SI06", "중독 문제"),
    ("SI10", "인지 기능 관련 어려움"),
    ("SI11", "민감한 피해 경험")
]

# 텍스트 생성 함수
def generate_fast_text():
    return (
        random.choice(long_starters)
        + random.choice(long_middles)
        + random.choice(long_endings)
    )

# 실제 생성 루프
generated_data = []
seen_texts = set()

# 진행률 표시용
target_total = 400
category_index = 0

while len(generated_data) < target_total:
    si_code, si_kor = situation_categories[category_index % 4]
    text = generate_fast_text()
    if len(text) >= 200 and text not in seen_texts:
        seen_texts.add(text)
        em_code, em_kor = random.choice(emotion_pool)
        generated_data.append({
            "text": text,
            "em_code": em_code,
            "em_kor": em_kor,
            "si_code": si_code,
            "si_kor": si_kor
        })
        # 진행상황 출력
        if len(generated_data) % 20 == 0:
            print(f"✅ {len(generated_data)}/{target_total} 생성 완료")
    category_index += 1


# 데이터프레임으로 저장
df = pd.DataFrame(generated_data)
df.to_csv("synthetic_consultation_clean_400.csv", index=False, encoding='utf-8-sig')
print("✅ 저장 완료: synthetic_consultation_clean_400.csv")
```
