````python
y = ax + b
a : 기울기(계수) coefficient -> coef_
b : 절편 intercept -> intercept_
````

```

X = 4
```

````python
from sklearn.linear_model import LinearRegression
X = [[4]]
y = [10]

lr = LinearRegression()
lr.fit(X, y)
````

````python
lr.coef_ # 기울기 a의 값 
````

````python
lr.intercept_ #  절편 b의 값 
````

```

```

````python
lr.predict([[100]])
````

```
12 = a*5 + b
8 = a*12 + b
```

````python
X = [[5], [12]]
y = [12, 8]

lr = LinearRegression()
lr.fit(X, y)
````

````python
a = lr.coef_ # 기울기 a의 값 
````

````python
b = lr.intercept_ #  절편 b의 값 
````

````python
a * 5 + b
````

````python
a * 12 + b
````

````python
lr.score(X, y) #  모델의 정확도 
````

````python
학습을 통하여 기울기와 절편을 찾아 직선의 방정식을 도출해낸다
여러 산점도를 통하여 산점도에 가장 이상적인, 수렴할 수 있는 직선의 방정식을 계산한다 
````

````python
coef = lr.coef_
intercept = lr.intercept_
score = lr.score(X, y)

print(f"y = {coef.round(2)}X + {intercept:.2f}")

````

````python
X = [[5], [12], [9]] #  벡터형식
y = [12, 8, 14]
lr = LinearRegression()
lr.fit(X, y)
````

````python
coef = lr.coef_
intercept = lr.intercept_
score = lr.score(X, y)
````

````python
print(coef, intercept, score)
````

````python
import matplotlib.pyplot as plt
plt.scatter(X, y, color='blue', marker='D')
y_pred = lr.predict(X)
plt.show()
````

````python
오차함수 - 모델의 정확도을 정량화 

실제 {1,2,3}
예상 {1,4,1}
오차값을 다 더했을 때 0이 나오게 된다 
절대값으로 더했을 때 4가 나옴(평균 1.3)


모델의 정확도를 평가하는 오차 
평균 절대 오차(Mean Absolute Error, MAE)



평균 제곱 오차(Mean square error, MSE)


평균 제곱근 오차 (RMSE)

R-square(R2제곱) : sklearn LinearRegression 모델의 score 메소드가 사용하는 척도 (결정계수)
1에 가까우면 완벽, 0이면 모델의 어긋남 
R제곱 = 예측분산/분산 
````
