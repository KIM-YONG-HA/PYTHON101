## **GridSearchCV**
최적의 매개변수값 조회

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/wine.csv')
```

```python
df.head()
```

```python
data = df.drop(columns=['class']).to_numpy()
target = df['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
sub_input, val_input, sub_target, val_target = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
# dt = DecisionTreeClassifier(random_state=42)
# dt.fit(sub_input, sub_target)
```

```python
from sklearn.model_selection import GridSearchCV
params = {'min_impurity_decrease':[0.0001, 0.0002, 0.0003, 0.0004, 0.0005]} #  최소 불순도
gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)

# 결정 트리(Decision Tree)의 노드 분할을 제어하는 하이퍼파라미터
```

```python
gs.fit(X_train, y_train)
```

```python
dt = gs.best_estimator_ #  가장 뛰어난 추정기(최적의 파라미터로 훈련된 모델이 들어있음.)
print(dt)
```

```python
print(gs.best_params_)
```

```python
gs.cv_results_
```

```python
print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```

```python
min_impurity_decrease = np.arange(0.0001, 0.001, 0.0001)
max_depth = range(5, 20, 1)
min_samples_split = range(2, 100, 10)
params = {
    'min_impurity_decrease': min_impurity_decrease, 
    'max_depth' :max_depth, 
    'min_samples_split': min_samples_split
} 

gs = GridSearchCV(DecisionTreeClassifier(random_state=42), params, n_jobs=-1)
gs.fit(X_train, y_train)
```

```python
dt = gs.best_estimator_
print(dt)
```

```python
print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```
