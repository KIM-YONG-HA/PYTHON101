```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# iris = sns.load_dataset('iris')
# iris.info()
```

```python
iris = load_iris()
#ris
```

```python
iris_data = iris.data
iris_target = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_target, random_state=42, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8))
plot_tree(dt, filled=True, feature_names=iris.feature_names)
plt.show()
```

```python
dt.feature_importances_
```

```python
test_sample = [[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,2.5,4.6,1.9],[3,2,1,0.2]]
dt.predict(test_sample) #  ['setosa', 'versicolor', 'virginica']
```
