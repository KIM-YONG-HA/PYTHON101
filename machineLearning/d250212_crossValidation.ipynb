## **교차검증(Cross Validation, CV)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/wine.csv')
```

```python
df.head()
```

```python
data = df.drop(columns=['class']).to_numpy()
target = df['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
sub_input, val_input, sub_target, val_target = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
print(sub_input.shape, val_input.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(sub_input, sub_target)
```

```python
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold

scores = cross_validate(dt, X_train, y_train, return_train_score=True) 
# 5개의 kfold로 검증된 test_score의 결과 
```

```python
train_score_mean = scores['train_score'].mean()
test_score_mean = scores['test_score'].mean()
print(train_score_mean)
print(test_score_mean)
```

```python
scores = cross_validate(dt, X_train, y_train, return_train_score=True, cv=StratifiedKFold()) #  분할기 지정 cv= 
```

```python
print(train_score_mean)
print(test_score_mean)
```

```python
splitter = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
scores = cross_validate(dt, X_train, y_train, return_train_score=True, cv=splitter) #  분할기 지정 cv= 
print(train_score_mean)
print(test_score_mean)
scores['test_score']
```
