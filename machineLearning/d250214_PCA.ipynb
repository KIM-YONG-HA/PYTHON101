## **주성분 분석**

#### 차원과 차원축소
- 데이터가 가진 속성을 특성이라 불렀다.
- 과일 사진의 경우 10000 개의 픽셀이 있기때문에 10000 개의 특성이 있는 셈
- 머신러닝에서는 이런 특성을 차원(dimension)이라고도 부른다.
- 10000 개의 특성은 결국 10000 개의 차원이라는 건데 이 차원을 줄일 수 있다면 저장 공간을 크게 절약 가능

- 2차원 배열과 1차원 배열의 차원은 조금 다르게 사용됨
- 다차원 배열에서 차원은 배열의 축 개수, 즉 2차원 배열에서는 행과 열이 차원
- 1차원 배열, 즉 벡터일 경우에는 원소의 개수를 말한다.

- 차원 축소는 데이터를 가장 잘 나타내는 일부 특성을 선택하여 데이터 크기를 줄이고 지도 학습 모델의 성능을 향상시킬 수 있는 방법
- 또한 줄어든 차원에서 다시 원본 차원으로 손실을 최대한 줄이면서 복원할 수도 있다.
- 대표적인 차원 축소 알고리즘인 주성분 분석(principal component analysis)를 학습, 주성분 분석을 PCA라고도 함

#### 주성분 분석

- 데이터에 있는 **분산이 큰 방향**을 찾는 것으로 이해(넓게 퍼진 벡터)
- 분산은 데이터가 널리 퍼져있는 정도를 말한다.
- 분산이 큰 방향을 데이터로 잘 표현하는 벡터로 생각할 수 있다.

주성분 분석(PCA) 후 수직으로 투영 => 차원 축소 



```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fruits = np.load('data/fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100 * 100)
```

```python

def draw_fruits(arr, ratio=1):
    n = len(arr)    # 샘플 개수
    
    rows = int(np.ceil(n/10))
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n 개까지만 그린다.
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
```

### PCA 클래스

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50, random_state=42)
pca.fit(fruits_2d)
```

```python
pca.components_.shape
```

```python
fruits_2d.shape
```

```python
pca.n_components_
```

```python
draw_fruits(fruits_2d.reshape(-1, 100, 100), ratio=0.5)
```

```python
draw_fruits(pca.components_.reshape(-1, 100, 100), ratio=1)
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape) # 300면, 10000개 픽셀값 -> 50개의 대표이미지, 10000개의 픽셀을 대표할만한 주성분 50개 
```

- 원본 데이터 재구성

```python
print(fruits_pca[0])
print(fruits_pca[1])
print(fruits_pca[2])
```

```python
fruits_inverse = pca.inverse_transform(fruits_pca)
print(fruits_inverse.shape)
```

```python
# fruits_re = fruits_inverse.reshape(-1, 100, 100)
# for start in [0, 100, 200]:
#     draw_fruits(fruits_re[start:start+100], ratio=0.5)

draw_fruits(fruits_inverse.reshape(-1, 100, 100), ratio=0.5)
```

#### 설명된 분산
- 주성분이 원본 데이터의 분산을 얼마나 잘 나타내는지 기록한 값을 설명된 분산(explained variance)
- PCA 클래스의 explained_variance_ratio_ 에 각 주성분의 설명된 분산 비율이 기록되어 있음

```python
print(pca.explained_variance_ratio_, end='\n\n')
print(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(pca.explained_variance_ratio_)
plt.show()
```

주성분이 5개 정도여도 괜찮은 성능이 나올 수 있다 

#### 다른 알고리즘과 함께 사용하기
- 과일 사진 원본 데이터와 PCA로 축소한 데이터를 지도학습에 적용해 보고 어떤 차이가 있는지 확인해보자
- 로지스틱 회귀를 사용

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
target = np.array([0] * 100 + [1] * 100 + [2] * 100)
```

- 로지스틱 회귀 모델에서 성능을 가믄해 보기 위해 cross_validate() 로 교차 검증 수행

```python
from sklearn.model_selection import cross_validate
scores = cross_validate(lr, fruits_2d, target, return_train_score=True)
```

```python
scores
```

```python
print(scores['fit_time'])
print(scores['test_score'])
```

```python
print(scores['fit_time'].mean())
print(scores['test_score'].mean())
```

```python
scores = cross_validate(lr, fruits_pca, target, return_train_score=True)
print(scores['fit_time'])
print(scores['test_score'])
```

```python
print(scores['fit_time'].mean())
print(scores['test_score'].mean())
```

```python
pca = PCA(n_components=0.5) # 원본데이터 50%를 설명하는 컴포넌트 
pca.fit(fruits_2d)
```

```python
pca.n_components_ #  2개의 차원만 있으면 50%
```

```python
fruits_pca = pca.transform(fruits_2d)
print(fruits_pca.shape)
```

```python
scores = cross_validate(lr, fruits_pca, target, return_train_score=True)
print(scores['fit_time'].mean())
print(scores['test_score'].mean())
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_pca)
print(np.unique(km.labels_, return_counts=True))
```

```python
for i in range(0, 3):
    draw_fruits(fruits[km.labels_ == i], ratio=0.5)
```

```python
#print(fruits[km.labels_ == 0])
for i in range(0, 3):
    data = fruits_pca[km.labels_ == i]
    plt.scatter(data[:, 0], data[:, 1], alpha=0.7)
plt.legend(['pineapple', 'banana', 'apple'])
plt.show()
```
