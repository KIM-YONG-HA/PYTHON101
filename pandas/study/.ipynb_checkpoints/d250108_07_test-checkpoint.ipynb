```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#  1. 남북한발전전력량.xlsx 파일 로드
excel = pd.read_excel('data/남북한발전전력량.xlsx')
excel
```

```python
#  2. 북한의 데이터만 df 저장
df = pd.DataFrame(excel[5:])
df
```

```python
#  3. 전력량(억kWh) 열 삭제
df.drop(columns=['전력량 (억㎾h)'], axis=1, inplace=True)
df
```

```python
#  4. 발전전력량 열을 인덱스로 지정 df 전치행렬
df.set_index('발전 전력별', inplace=True)
df = df.T
```

```python
#  5. 원자력 열의 '-'을 '0', 실수형으로 변환
df['원자력'] = df['원자력'].replace("-",0)
```

```python
df
```

```python
# data = df_tran.iloc[1:, :]
# df_new = pd.DataFrame(data)				
# df_new.columns = ['합계','수력','화력','원자력']
# df_new['원자력'] = df_new['원자력'].astype(float)
df['원자력'] = df['원자력'].astype(float)
df.info()
```

```python
#  6. 합계 열을 총 발전량으로 변경
df.rename(columns={'합계':'총 발전량'}, inplace=True)
df.head()
```

```python
#  7. 총발전량의 증감률을 계산해서 증감률 열 생성
df['총 발전량 - 1년'] = df['총 발전량'].shift(1)
```

```python
df
```

```python
# 100 200

(200-100)/200 * 100
```

```python
# (최종 – 최초) / 최초 × 100
#(263 - 277) / 263 * 100

#  현재값 대비 
#df['증감률'] = ((df['총 발전량'] - df['총 발전량 - 1년']) / df['총 발전량']) * 100

#  전년 대비 증감률 
df['증감률'] = ((df['총 발전량'] / df['총 발전량 - 1년']) - 1) * 100
```

```python
#   8. 이중 축 혼합형 차트 

from matplotlib import font_manager, rc

font_path = "data/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.rcParams['axes.unicode_minus'] = False

ax1 = df[['수력', '화력']].plot(kind='bar', figsize=(20, 10), width=0.7, stacked=True)
ax2 = ax1.twinx()

ax2.plot(df.index, df.증감률, ls='--', marker='o', markersize=10, color='#333', label='전년도 대비 증감률(%)')

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량', size=20)
ax2.set_ylabel('전년도 대비 증감률(%)', size=20)

plt.title('북한 전력 발전량(1990~2016)', size=25)

ax1.legend(loc='upper left')

plt.show()
```
