```python
import pandas as pd
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
columns = ['mpg', 'cylinders', 'displayment', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
df = pd.read_csv('data/auto-mpg.csv', header=None, names=columns)
df.head()
```

```python
cylinders_size = (df['cylinders'] / df['cylinders'].max()) * 200 
```

```python
s = pd.Series(cylinders_size)
```

```python
df.plot(kind='scatter', x='weight', y='mpg', c='#fa1', figsize=(10, 5), marker='o', s=cylinders_size, alpha=0.3)
plt.title('mpg-weight-cylinders')
plt.savefig('data/scatter.png')
#plt.savefig('data/scatter.png', transparent=True)
plt.show()
```

```python
df['count'] = 1
```

```python
df_origin = df.groupby(by='origin').sum(numeric_only=True)
df_origin.head()
```

```python
df_origin.index = ['USA', 'EU', 'JPN']
df_origin
```

```python
#count = df_origin['mpg']

# from matplotlib import font_manager, rc
# font_path = 'data/malgul.ttf'
# font_name = font_manager()


df_origin['count'].plot(kind='pie', figsize=(5, 5), autopct='%1.4f%%', color=['red', 'orange', 'skyblue'])

plt.title('Model Origin Count', size=20)
plt.legend(labels=df_origin.index, loc='upper right')

plt.show()
```
