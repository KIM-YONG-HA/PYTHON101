# Boolean indexing

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
index_bool = [False] * 15
index_bool[3] = True
index_bool[6] = True

df_bool = titanic.loc[:, index_bool]
df_bool.head()
```

```python
## 평균값이 30이상인 컬럼만 인덱싱(필터링)
```

```python
avg_bool = df_bool.apply(lambda x: x.mean(numeric_only=True) >= 30, axis=0)
#avg_bool

result = df_bool.loc[:, avg_bool]
result.head()

```
