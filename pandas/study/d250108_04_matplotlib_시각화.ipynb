```python
1. 시도별_전출입_인구수.xlsx 로드
2. 전출지별 열 누락데이터 ffill()
3. 전출지별 열에서 서울특별시 값을 갖는 데이터만 출력 
    - 전출지가 서울특별시, 전입지가 서울특별시가 아닌 데이터
    - df_seoul에 저장
4. 전입지별 열의 이름을 전입지로 변경
5. 전입지 열을 df_seoul의 인덱스로 지정
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('data/시도별_전출입_인구수.xlsx')
df.head()
```

```python
df['전출지별'] = df['전출지별'].ffill()
```

```python
df.head()
```

3. 전출지별 열에서 서울특별시 값을 갖는 데이터만 출력 
    - 전출지가 서울특별시, 전입지가 서울특별시가 아닌 데이터
    - df_seoul에 저장

```python
cond = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[cond].copy()
df_seoul.head()
```

```python
df_seoul.drop('전출지별', axis=1, inplace=True)
```

4. 전입지별 열의 이름을 전입지로 변경

```python
df_seoul.rename(columns={'전입지별':'전입지'}, inplace=True)
```

```python
df_seoul.head()
```

5. 전입지 열을 df_seoul의 인덱스로 지정

```python
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head()
```

```python
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

### annotate() 옵션

text (첫 번째 인자)

주어진 코드에서는 ''로 빈 문자열을 사용했습니다. 이 값은 텍스트를 추가하려면 문자열을 넣으면 됩니다.
예: plt.annotate('특정 위치', ...)처럼 사용할 수 있습니다.
xy

화살표의 끝점 좌표를 설정합니다.
여기서 (47, 450000)은 화살표가 가리키는 지점입니다.
xytext

화살표의 시작점 좌표를 설정합니다.
여기서 (30, 620000)은 화살표가 시작하는 지점입니다.
xycoords

xy와 xytext가 어떤 좌표 체계를 따르는지 설정합니다.
기본값은 'data'로, 데이터 값 기반 좌표를 사용합니다.
다른 예:
'axes fraction': 축 내부에서 좌표를 비율로 표현 (0.0 ~ 1.0).
'figure fraction': 전체 그림의 비율 좌표 (0.0 ~ 1.0).
'offset points': 지정된 점에서의 픽셀 단위.
arrowprops

화살표의 스타일과 속성을 지정하는 딕셔너리입니다.
arrowprops의 키와 의미
arrowstyle

화살표의 모양을 정의합니다.
주요 스타일:
'->': 기본 화살표.
'-[': 직선형 화살표.
'|-|', 'fancy': 다양한 디자인 제공.
color

화살표의 색상을 지정합니다.
여기서는 'red'로 설정되어 있습니다.
lw (line width)

화살표 선의 두께를 설정합니다.
여기서는 5로 설정되어 화살표가 두껍게 표시됩니다.
기타 옵션

shrinkA / shrinkB: 화살표 시작점과 끝점에서 화살표 길이를 조정합니다.
mutation_scale: 화살표 모양의 크기를 조정합니다.
connectionstyle: 화살표 연결선을 설정합니다 (예: 'arc3').

```python
plt.figure(figsize=(14,5))
plt.xticks(rotation=-45) ## 반시계 방향 default
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=7)
plt.title('서울 -> 경기 전출')
plt.xlabel('연도')
plt.ylabel('이동 인구수')
plt.legend(labels=['서울-경기'], loc='best')
plt.ylim(50000, 800000)

plt.annotate('',  xytext=(2, 290000), xy=(20, 620000), xycoords='data', arrowprops=dict(arrowstyle='->', color='skyblue', lw=5))

plt.annotate('',  xytext=(30, 620000), xy=(47, 450000), xycoords='data', arrowprops=dict(arrowstyle='->', color='red', lw=5))

#plt.annotate('상승', xy=(20, 620000), xytext=(2, 290000), xycoords='data', arrowprops=dict(arrowstyle='->', color='skyblue', lw=5))

plt.show()

```
