```python
3. 주가 데이터.xlsx 파일로 df 만들고 연월일 컬럼을 연 월 일 3개로 분할
연월일 컬럼은 str타입으로 변경 후 -로 분할
```

```python
df = pd.read_excel('data/주가데이터.xlsx')
df['연월일'].info()
```

```python
df['연월일'] = df['연월일'].astype(str)
```

```python
df['연월일'].info()
```

```python
count = int(len(df['연월일'].values))
arr = df['연월일'].values

for i in range(count):
    df['연'] = arr[i].split("-")[0]
    df['월'] = arr[i].split("-")[1]
    df['일'] = arr[i].split("-")[2]
```
