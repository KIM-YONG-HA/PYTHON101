# Boolean indexing

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
index_bool = [False] * 15
index_bool[3] = True
index_bool[6] = True

df_bool = titanic.loc[:, index_bool]
df_bool.head()
```

```python
## 평균값이 30이상인 컬럼만 인덱싱(필터링)
```

```python
avg_bool = df_bool.apply(lambda x: x.mean(numeric_only=True) >= 30, axis=0)
#avg_bool

result = df_bool.loc[:, avg_bool]
result.head()

```

```python
## 각 행의  평균값이 50 이상인 이면 yes, 아니면 no, result 열에 저장 
```

```python
df = titanic.loc[:, ['age', 'fare']]
df_copy = df.copy()
```

```python
# def test(a):
#     print(type(a))
#     print(a)
#     return a

# df_copy.apply(test, axis=1)
```

```python
df_copy['mean'] = df_copy.apply(lambda x: x.mean(), axis=1)
```

```python
df_copy['result'] = df_copy.apply(lambda x: 'Yes' if x.mean() >= 50 else 'No', axis=1)
```
