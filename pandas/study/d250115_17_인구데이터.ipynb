```python

import matplotlib.pyplot as plt

title = ''
years = []
values = []
loc_name = input("지역 입력 : ")

for key, value in dic.items():
    if key.find(loc_name) != -1:  # '용봉'이 포함된 키를 찾음
        
        print("key :", key)
        # title = key
        
        for i in range(len(value)):
            # 연도 추출 및 출력
            year = value.index[i][:4]  # 인덱스에서 연도 추출
            data = value.iloc[i]
            #print(f"{year}년 :", data)  # iloc로 값 출력
            years.append(year)
            values.append(data)


df = pd.DataFrame(values, index=years, columns=["인구수"])
df['인구수'] = df['인구수'].fillna(0)
df['인구수'] = df['인구수'].apply( lambda x: int(x.replace(",", "")) if isinstance(x, str) or x == "0" else x)

# df['인구수'] = df['인구수'].astype(int)
df['인구수'].plot(kind='bar', figsize=(7, 4))


```
