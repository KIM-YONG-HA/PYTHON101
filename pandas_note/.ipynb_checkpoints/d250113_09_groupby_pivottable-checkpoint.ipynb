```python
# groupby 
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
gender = df.groupby('sex').mean(numeric_only=True)
gender
```

```python
df.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean()
```

```python
df.groupby(['sex', 'pclass'])[['survived']].mean()
```

```python
group_data = df.groupby(['sex', 'pclass'])['survived'].mean()
data_mean = pd.DataFrame(group_data)
data_mean
```

```python
## 인덱스 초기화 
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean().reset_index()
```

```python
## 다중컬럼 결과 출력
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].mean()
```

```python
## 다중 통계 함수 적용
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

```python
## agg 딕셔너리 사용 
```

```python
df.groupby(['sex', 'pclass']).agg({'survived':['mean','median'], 'age':['mean','median']})
```

```python
## 사용자 정의 함수 사용 
```

```python
def max_min(x):
    return x.max() - x.min()

df.groupby(['sex', 'pclass'])[['fare', 'age']].agg(max_min)
```
