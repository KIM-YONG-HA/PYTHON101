```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
sample = df.copy()
sample.head()
```

```python
sample['class'].value_counts()
```

sample 데이터프레임에 대하여 apply()를 활용하여 class 컬럼의 값을 다음과 같이 바꾸고, 
분포를 출력후 변경 전과 동일한지 확인하세요

삼등석    491
일등석    216
이등석    184
Name: class, dtype: int64

```python
def transform_class(x):
    if x == 'First':
        return '일등석'
    elif x == 'Second':
        return '이등석'
    else:
        return '삼등석'


sample['class'] = sample['class'].apply(func=transform_class)
#sample.head()
```

```python
sample['class'].value_counts()
```
