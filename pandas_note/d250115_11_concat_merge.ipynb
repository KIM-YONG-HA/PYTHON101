````python
# merge

공통된 컬럼(key값)을 가지고 있다면 병합할 수 있다.
````

````python
df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
df1
````

````python
df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
df2
````

````python
pd.merge(df1, df2) #  inner 옵션
````

````python
## how 옵션 : 병합 방식
  
left : 왼쪽 데이터프레임 행을 유지하고 일치하는 오른쪽 데이터를 병합(값이 없다면 NaN)
right : 오른쪽 데이터프레임 행을 유지하고 일치하는 왼쪽 데이터를 병합(값이 없다면 NaN)
outer : 두 데이터프레임의 행을 유지하고 병합하며 값이 없다면 NaN
inner : 공통된 키값(행)만 병합한다.

````

````python
### left
````

````python
pd.merge(df1, df2, how='left')
````

````python
### right
````

````python
pd.merge(df1, df2, how='right')
````

````python
### outer
````

````python
pd.merge(df1, df2, how='outer')
````

````python
### inner
````

````python
pd.merge(df1, df2, how='inner')
````
