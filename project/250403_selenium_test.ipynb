```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
import pyautogui

# 옵션 설정
options = Options()
options.add_argument("--start-maximized")  # 창 최대화
# options.add_argument("--headless")  # 시연이므로 주석 처리 (UI 보이게)

# 크롬 브라우저 새 창으로 열기
driver = webdriver.Chrome(options=options)
url = "http://vb901217.dothome.co.kr/maumcheck/index.php"
driver.get(url)
time.sleep(3)




# 스크롤 설정
scroll_pause_time = 0.05  # 스크롤 간 대기 시간
scroll_step = 50          # 한 번에 이동할 픽셀 수

# 페이지 전체 높이
total_height = driver.execute_script("return document.body.scrollHeight")

# ▼ 아래로 천천히 스크롤
current_scroll = 0
while current_scroll < total_height:
    driver.execute_script(f"window.scrollBy(0, {scroll_step});")
    current_scroll += scroll_step
    time.sleep(scroll_pause_time)

# 도달 후 잠깐 대기
time.sleep(1)

# ▲ 위로 천천히 스크롤
while current_scroll > 0:
    driver.execute_script(f"window.scrollBy(0, -{scroll_step});")
    current_scroll -= scroll_step
    time.sleep(scroll_pause_time)

# 상단 도달 후 잠깐 대기
time.sleep(2)




# 원하는 위치로 이동 (x, y는 화면 좌표)
pyautogui.moveTo(800, 500, duration=1.5)  # 1.5초 동안 부드럽게 이동
pyautogui.click()




# 글로 작성하기 
write_btn = driver.find_element(By.XPATH, '//*[@id="visual"]/div[5]/div/a[1]')
time.sleep(2)

# 버튼 클릭
#write_btn.click()









# 종료는 수동으로 하거나 아래 코드로 자동 종료 가능
#driver.quit()




```
