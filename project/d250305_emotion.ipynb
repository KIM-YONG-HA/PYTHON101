````python
import MeCab

# MeCab 실행 (설치된 사전 경로 확인 후 조정 가능)
mecab = MeCab.Tagger("-r C:/mecab/etc/mecabrc")

# 불용어 리스트 (필요하면 확장 가능)
stopwords = korean_stopwords

# 형태소 분석 후 불필요한 품사 제거 및 원형 변환 적용
def preprocess_text(text):
    if not isinstance(text, str):  # NaN 값 방지
        return ""

    parsed_text = mecab.parse(text).split("\n")[:-2]  # 마지막 줄(EOS, 공백) 제거
    tokens = [line.split("\t") for line in parsed_text if "\t" in line]  # 탭 기준으로 나누기

    # 품사 필터링 (명사, 동사, 형용사만 유지)
    keywords = []
    for token in tokens:
        if len(token) < 2:
            continue
        word, features = token[0], token[1].split(",")  # 형태소와 품사 정보 분리
        tag = features[0]  # 품사 태그 추출
        lemma = features[-3]  # 원형(기본형) 추출 (활용형이 있는 경우)

        # 🔹 명사(NNG, NNP) 그대로 추가
        if tag in ['NNG', 'NNP'] or word == "키":
            keywords.append(word)

        # 🔹 동사(VV), 형용사(VA) 원형 변환
        elif tag in ['VA', 'VV']:
            if lemma != '*' and lemma != '' and lemma != word:  # 원형 정보가 있는 경우
                keywords.append(lemma)
            else:
                keywords.append(word + "다")  # 원형이 없으면 '다'를 붙여 변환

    # 🛑 VV, VA 태그가 포함된 경우 최종적으로 제거
    keywords = [word for word in keywords if word not in ['VV', 'VA']]

    # 중복 단어 제거 & 불용어 제거 (단, "키"는 유지)
    keywords = list(dict.fromkeys(keywords))  # 중복 제거
    keywords = [word for word in keywords if word not in stopwords or word == "키"]  # 불용어 제거 (키 제외)

    return " ".join(keywords)

# MeCab 적용하여 텍스트 전처리
df_cleaned['processed_text'] = df_cleaned['HS01'].dropna().apply(preprocess_text)

# 결과 확인 (VV, VA가 제거되었는지 확인)
print(df_cleaned.loc[45767, 'processed_text'])
print(df_cleaned.loc[16549, 'processed_text'])
print(df_cleaned.loc[12367, 'processed_text'])
print(df_cleaned.loc[45502, 'processed_text'])
````
